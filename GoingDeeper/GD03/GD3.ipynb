## Build For CAM Model

Res-Net 50 을 기반으로 만든 CAM 모델.
pooling layer 뒤에 soft-max를 붙여서 grad-CAM 모델을 생성. 

reference: aiffel 노드 8-4. CAM, Grad-CAM용 모델 준비하기. 

```python
num_classes = ds_info.features["label"].num_classes
base_model = keras.applications.resnet50.ResNet50(
    include_top=False,     # Imagenet 분류기  fully connected layer 제거
    weights='imagenet',
    input_shape=(224, 224,3),
    pooling='avg',      # GAP를 적용  
)
x = base_model.output
preds = keras.layers.Dense(num_classes, activation='softmax')(x)
cam_model = keras.Model(inputs=base_model.input, outputs=preds)
```

```python
cam_model.summary()
```

# Training CAM Model

1. 모델 normalize 와 resizing 함수를 이용하여 전처리 후 학습.
2. apply_normalize_on_dataset() 함수를 사용하여 배치 구성. 

```python
def normalize_and_resize_img(input):
    # Normalizes images: `uint8` -> `float32`
    image = tf.image.resize(input['image'], [224, 224])
    input['image'] = tf.cast(image, tf.float32) / 255.
    return input['image'], input['label']

def apply_normalize_on_dataset(ds, is_test=False, batch_size=16):
    ds = ds.map(
        normalize_and_resize_img, 
        num_parallel_calls=2
    )
    ds = ds.batch(batch_size)
    if not is_test:
        ds = ds.repeat()
        ds = ds.shuffle(200)
    ds = ds.prefetch(tf.data.experimental.AUTOTUNE)
    return ds
```

```python
# 데이터셋에 전처리와 배치처리를 적용합니다. 
ds_train_norm = apply_normalize_on_dataset(ds_train)
ds_test_norm = apply_normalize_on_dataset(ds_test)

# 구성된 배치의 모양을 확인해 봅니다. 
for input in ds_train_norm.take(1):
    image, label = input
    print(image.shape)
    print(label.shape)
```

```python
tf.random.set_seed(2021) #가중치를 무작위로 초기화 시켜서 결과에대한 재현성을 확보하기 위해서 매번 실행해도 동일한 weight 초기화, dropout mask, 배치 순서 등을 사용하여 결과를 동일하게 재현.
cam_model.compile(
    loss='sparse_categorical_crossentropy',
    optimizer=tf.keras.optimizers.SGD(lr=0.01),
    metrics=['accuracy'],
)
```

```python
history_cam_model = cam_model.fit(
    ds_train_norm,
    steps_per_epoch=int(ds_info.splits['train'].num_examples/16),
    validation_steps=int(ds_info.splits['test'].num_examples/16),
    epochs=20,
    validation_data=ds_test_norm,
    verbose=1,
    use_multiprocessing=True,
)
```

총 20 epochs 수 중에서 11 epochs 까지 학습을 시켰을 때의 loss가 0.0053 과 accuracy는 1.00을 달성하였고, validation_loss = 0.97 그리고 validation_accuracy = 0.74 정도로 나왔습니다. 따라서 이에대한 해석으로는 모델이 예측한 값과 실제 값 사이의 차이를 보면 손실값이 매우 작고, 훈련 데이터셋에 대한 모델 예측도가 100%의 정확도를 보여주면서 모델이 훈련 데이터를 잘 학습을 했지만, 검증 데이터셋에 대한 손실값을 보면 훈련 데이터 손실에 비해 매우 높은 확률 로 성능을 일반화 할 수 없다는 것을 시사합니다. 또한 모델이 처음 보는 데이터에 대해서는 검증 데이터의 정확도를 보면 74% 정도의 정확도를 확인할 수 있습니다. 이 결과를 종합적으로 확인 했을 때, 모델의 학습 상황이 과적합이 일어났을 경우가 매우 높습니다. 이는 모델이 훈련 데이터의 특정 패턴이나 노이즈까지 과도하게 학습하여 새로운 데이터 셋에 대해서는 좋은 성능을 내지 못하는것을 의미하며 훈련 데이터를 통해서 '암기'는 잘하지만, 일반적인 규칙은 학습하지 못하는 상태 이기 때문에, 과적합을 줄이기 위해서 규제 적용, 드롭아웃, 혹은 많은 데이터 확보, 조기 종료 등을 고려 해 볼 필요가 있을것 같습니다.

# Training CAM 모델 전제 재구성

```python
from tensorflow.keras.applications.resnet50 import preprocess_input

#앞에 학습된 가중치를 가진 모델이 memory에 남아 있기 때문에 모델을 다시 정의를 해줘서 초기화 시키고 모델 학습.
num_classes = ds_info.features["label"].num_classes
base_model = keras.applications.resnet50.ResNet50(
    include_top=False,     # Imagenet 분류기  fully connected layer 제거
    weights='imagenet',
    input_shape=(224, 224,3),
    pooling='avg',      # GAP를 적용  
)
x = base_model.output
preds = keras.layers.Dense(num_classes, activation='softmax')(x)
cam_model = keras.Model(inputs=base_model.input, outputs=preds)

def normalize_and_resize_img(input_dict): # 입력 인자 이름을 input_dict 등으로 명확히 하는 것이 좋습니다.
    image = tf.image.resize(input_dict['image'], [224, 224])
    image = tf.cast(image, tf.float32)
    image = preprocess_input(image)  # ResNet50 전용 preprocess_input 사용!
    return image, input_dict['label']

def apply_normalize_on_dataset(ds, is_test=False, batch_size=16):
    ds = ds.map(
        normalize_and_resize_img,
        num_parallel_calls=2
    )
    ds = ds.batch(batch_size) # 1. 배치 만들기
    if not is_test:          # 2. 훈련 데이터셋 (is_test=False) 인 경우에만!
        ds = ds.repeat()     # 3. 데이터셋 반복
        ds = ds.shuffle(200) # 4. <<< 여기에 셔플이 적용됩니다! (버퍼 크기 200)
    ds = ds.prefetch(tf.data.experimental.AUTOTUNE)
    return ds

# 함수 호출
# 훈련 데이터셋: is_test=False 이므로 셔플 적용됨
ds_train_norm = apply_normalize_on_dataset(ds_train, is_test=False, batch_size=16)
# 테스트/검증 데이터셋: is_test=True 이므로 셔플 적용 안 됨
ds_test_norm = apply_normalize_on_dataset(ds_test, is_test=True, batch_size=16)

# 구성된 배치의 모양을 확인해 봅니다. 
for input in ds_train_norm.take(1):
    image, label = input
    print(image.shape)
    print(label.shape)
```

```python
tf.random.set_seed(2021) #가중치를 무작위로 초기화 시켜서 결과에대한 재현성을 확보하기 위해서 매번 실행해도 동일한 weight 초기화, dropout mask, 배치 순서 등을 사용하여 결과를 동일하게 재현.
cam_model.compile(
    loss='sparse_categorical_crossentropy',
    optimizer=tf.keras.optimizers.SGD(lr=0.01),
    metrics=['accuracy'],
)
```

```python
# 일반 정규화 모델 - 커널 재시작으로 이전 스텝의 코드를 모아서 한꺼번에 실행합니다.
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
import tensorflow_datasets as tfds
import copy
import cv2
from PIL import Image
import os


(ds_train, ds_test), ds_info = tfds.load(
    'stanford_dogs',
    split=['train', 'test'],
    shuffle_files=True,
    with_info=True,
)

def normalize_and_resize_img(input):
    # Normalizes images: `uint8` -> `float32`
    image = tf.image.resize(input['image'], [224, 224])
    input['image'] = tf.cast(image, tf.float32) / 255.
    return input['image'], input['label']

def apply_normalize_on_dataset(ds, is_test=False, batch_size=16):
    ds = ds.map(
        normalize_and_resize_img, 
        num_parallel_calls=2
    )
    ds = ds.batch(batch_size)
    if not is_test:
        ds = ds.repeat()
        ds = ds.shuffle(200)
    ds = ds.prefetch(tf.data.experimental.AUTOTUNE)
    return ds
```

```python
# 사전학습 기반 정규화 모델 - 커널 재시작으로 이전 스텝의 코드를 모아서 한꺼번에 실행합니다.
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
import tensorflow_datasets as tfds
import copy
import cv2
from PIL import Image
import os

(ds_train, ds_test), ds_info = tfds.load(
    'stanford_dogs',
    split=['train', 'test'],
    shuffle_files=True,
    with_info=True,
)

def normalize_and_resize_img(input_dict): # 입력 인자 이름을 input_dict 등으로 명확히 하는 것이 좋습니다.
    image = tf.image.resize(input_dict['image'], [224, 224])
    image = tf.cast(image, tf.float32)
    image = preprocess_input(image)  # ResNet50 전용 preprocess_input 사용!
    return image, input_dict['label']

def apply_normalize_on_dataset(ds, is_test=False, batch_size=16):
    ds = ds.map(
        normalize_and_resize_img,
        num_parallel_calls=2
    )
    ds = ds.batch(batch_size) # 1. 배치 만들기
    if not is_test:          # 2. 훈련 데이터셋 (is_test=False) 인 경우에만!
        ds = ds.repeat()     # 3. 데이터셋 반복
        ds = ds.shuffle(200) # 4. <<< 여기에 셔플이 적용됩니다! (버퍼 크기 200)
    ds = ds.prefetch(tf.data.experimental.AUTOTUNE)
    return ds
```

```python
def generate_cam(model, item):
    item = copy.deepcopy(item)
    width = item['image'].shape[1]
    height = item['image'].shape[0]
    
    img_tensor, class_idx = normalize_and_resize_img(item)
    
    # 학습한 모델에서 원하는 Layer의 output을 얻기 위해서 모델의 input과 output을 새롭게 정의해줍니다.
    # model.layers[-3].output에서는 우리가 필요로 하는 GAP 이전 Convolution layer의 output을 얻을 수 있습니다.
    cam_model = tf.keras.models.Model([model.inputs], [model.layers[-3].output, model.output])
    conv_outputs, predictions = cam_model(tf.expand_dims(img_tensor, 0))
    
    conv_outputs = conv_outputs[0, :, :, :]
    class_weights = model.layers[-1].get_weights()[0] #마지막 모델의 weight activation을 가져옵니다.
    
    cam_image = np.zeros(dtype=np.float32, shape=conv_outputs.shape[0:2])
    for i, w in enumerate(class_weights[:, class_idx]):
        # W * f 를 통해 class별 activation map을 계산합니다.
        cam_image += w * conv_outputs[:, :, i]

    cam_image /= np.max(cam_image) # activation score를 normalize합니다.
    cam_image = cam_image.numpy()
    cam_image = cv2.resize(cam_image, (width, height)) # 원래 이미지의 크기로 resize합니다.
    return cam_image
```

```python
def visualize_cam_on_image(src1, src2, alpha=0.5):
    beta = (1.0 - alpha)
    merged_image = cv2.addWeighted(src1, alpha, src2, beta, 0.0)
    return merged_image
```

# Grad-CAM Implementation

만든 함수로 여러 레이어의 CAM 이미지를 뽑아서 직접 확인 하고, 위에서 만든 이미지 visualize_cam_on_image() 함수로 이미지를 시각화.

```python
def generate_grad_cam(model, activation_layer, item):
    item = copy.deepcopy(item)
    width = item['image'].shape[1]
    height = item['image'].shape[0]
    img_tensor, class_idx = normalize_and_resize_img(item)
    
    # Grad cam에서도 cam과 같이 특정 레이어의 output을 필요로 하므로 모델의 input과 output을 새롭게 정의합니다.
    # 이때 원하는 레이어가 다를 수 있으니 해당 레이어의 이름으로 찾은 후 output으로 추가합니다.
    grad_model = tf.keras.models.Model([model.inputs], [model.get_layer(activation_layer).output, model.output])
    
    # Gradient를 얻기 위해 tape를 사용합니다.
    with tf.GradientTape() as tape:
        conv_output, pred = grad_model(tf.expand_dims(img_tensor, 0))
    
        loss = pred[:, class_idx] # 원하는 class(여기서는 정답으로 활용) 예측값을 얻습니다.
        output = conv_output[0] # 원하는 layer의 output을 얻습니다.
        grad_val = tape.gradient(loss, conv_output)[0] # 예측값에 따른 Layer의 gradient를 얻습니다.

    weights = np.mean(grad_val, axis=(0, 1)) # gradient의 GAP으로 class별 weight를 구합니다.
    grad_cam_image = np.zeros(dtype=np.float32, shape=conv_output.shape[0:2])
    for k, w in enumerate(weights):
        # 각 class별 weight와 해당 layer의 output을 곱해 class activation map을 얻습니다.
        grad_cam_image += w * output[:, :, k]
        
    grad_cam_image /= np.max(grad_cam_image)
    grad_cam_image = grad_cam_image.numpy()
    grad_cam_image = cv2.resize(grad_cam_image, (width, height))
    return grad_cam_image
```

```python
def get_bbox(cam_image, score_thresh=0.05):
    low_indicies = cam_image <= score_thresh
    cam_image[low_indicies] = 0
    cam_image = (cam_image*255).astype(np.uint8)
    
    contours,_ = cv2.findContours(cam_image, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    cnt = contours[0]
    rotated_rect = cv2.minAreaRect(cnt)
    rect = cv2.boxPoints(rotated_rect)
    rect = np.int0(rect)
    return rect
```

```python
def get_bbox(cam_image_input, score_thresh=0.05, min_contour_area=50): # min_contour_area 추가
    # 원본 cam_image를 수정하지 않기 위해 복사본 사용
    cam_image = np.copy(cam_image_input)

    # CAM 이미지 값 분포 확인 (디버깅용)
    # print(f"Original CAM for BBox min: {cam_image.min()}, max: {cam_image.max()}, mean: {cam_image.mean()}")
    # plt.imshow(cam_image, cmap='jet')
    # plt.title("CAM before thresholding in get_bbox")
    # plt.colorbar()
    # plt.show()

    low_indicies = cam_image <= score_thresh
    cam_image[low_indicies] = 0

    # 임계값 처리 후 CAM 이미지 확인 (디버깅용)
    # plt.imshow(cam_image, cmap='jet')
    # plt.title(f"CAM after thresholding (thresh: {score_thresh}) in get_bbox")
    # plt.colorbar()
    # plt.show()

    # CAM 이미지가 완전히 0이 되었는지 확인
    if np.max(cam_image) == 0:
        print("Warning: CAM image is all zeros after thresholding. No bbox can be generated.")
        return np.array([[0,0], [0,0], [0,0], [0,0]], dtype=np.int0) # 빈 박스 또는 예외 처리

    cam_image_uint8 = (cam_image * 255).astype(np.uint8)

    contours, _ = cv2.findContours(cam_image_uint8, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE) # RETR_EXTERNAL 사용 추천

    if not contours:
        print("Warning: No contours found after thresholding.")
        return np.array([[0,0], [0,0], [0,0], [0,0]], dtype=np.int0) # 빈 박스 또는 예외 처리

    # 가장 큰 윤곽선 선택
    cnt = max(contours, key=cv2.contourArea)

    # 너무 작은 윤곽선은 무시 (선택 사항)
    if cv2.contourArea(cnt) < min_contour_area:
        print(f"Warning: Largest contour area {cv2.contourArea(cnt)} is less than min_contour_area {min_contour_area}.")
        return np.array([[0,0], [0,0], [0,0], [0,0]], dtype=np.int0)


    rotated_rect = cv2.minAreaRect(cnt)
    rect = cv2.boxPoints(rotated_rect)
    rect = np.int0(rect)
    return rect

# Grad-CAM 이미지로 바운딩 박스 생성 및 시각화 (수정된 get_bbox 사용)
grad_cam_image_for_bbox = generate_grad_cam(cam_model, 'conv5_block3_out', item) # 또는 다른 CAM 이미지 사용

# CAM 이미지 시각화 (get_bbox에 전달하기 전 상태 확인)
plt.imshow(grad_cam_image_for_bbox, cmap='jet')
plt.title("Grad-CAM for BBox Generation")
plt.colorbar()
plt.show()

image_for_grad_bbox = copy.deepcopy(item['image']).astype(np.uint8) # 원본 이미지 복사 및 타입 확인

# get_bbox 호출 시 임계값(score_thresh) 및 최소 면적(min_contour_area)을 조정해보세요.
# score_thresh를 낮추거나 (예: 0.01, 0.005), CAM 값의 분포를 보고 동적으로 설정할 수도 있습니다.
# 예: dynamic_thresh = np.percentile(grad_cam_image_for_bbox[grad_cam_image_for_bbox > 0], 10) # 0이 아닌 값들 중 하위 10%
grad_rect = get_bbox(grad_cam_image_for_bbox, score_thresh=0.1, min_contour_area=100) # 임계값 조정

if grad_rect.any(): # 바운딩 박스가 유효하게 생성된 경우에만 그리기
    image_with_grad_bbox = cv2.drawContours(image_for_grad_bbox, [grad_rect], 0, (0,0,255), 2)
    plt.imshow(image_with_grad_bbox)
    plt.title("Image with Grad-CAM BBox")
    plt.show()
else:
    print("Failed to generate a valid bounding box from Grad-CAM.")
    plt.imshow(image_for_grad_bbox) # 바운딩 박스 없이 원본 이미지 표시
    plt.title("Image (No BBox generated)")
    plt.show()
```

# IoU (Intersection over Union) 구하기

1. 정답 바운딩 박스(ground truth) 와 예측값 바운딩 박스(prediction box) 두 가지를 이용해서 IoU를 계산하는 get_iou() 함수를 만들어 주세요.

2. CAM과 Grad-CAM으로 얻은 바운딩 박스와 정답 바운딩 박스와의 IoU를 각각 비교해 봅시다.

```python
def rect_to_minmax(rect, image_shape_tuple): # image_shape_tuple은 (height, width, channels) 형태
    # rect가 유효한지 먼저 확인 (점 4개로 구성)
    if rect is None or not isinstance(rect, np.ndarray) or rect.shape != (4, 2):
        print("Warning: Invalid rect input to rect_to_minmax. Returning zero bbox.")
        return [0.0, 0.0, 0.0, 0.0] # 유효하지 않은 경우 빈/0 박스 반환

    img_height = float(image_shape_tuple[0])
    img_width = float(image_shape_tuple[1])

    # rect의 각 점 [x, y]에서 y좌표와 x좌표를 분리
    y_coordinates = rect[:, 1]
    x_coordinates = rect[:, 0]

    # 정규화 및 클리핑
    # y_min: 모든 y좌표 중 최소값 / 이미지 높이, 0과 1 사이로 클리핑
    y_min_norm = np.clip(np.min(y_coordinates) / img_height, 0.0, 1.0)
    # x_min: 모든 x좌표 중 최소값 / 이미지 너비, 0과 1 사이로 클리핑
    x_min_norm = np.clip(np.min(x_coordinates) / img_width, 0.0, 1.0)
    # y_max: 모든 y좌표 중 최대값 / 이미지 높이, 0과 1 사이로 클리핑
    y_max_norm = np.clip(np.max(y_coordinates) / img_height, 0.0, 1.0)
    # x_max: 모든 x좌표 중 최대값 / 이미지 너비, 0과 1 사이로 클리핑
    x_max_norm = np.clip(np.max(x_coordinates) / img_width, 0.0, 1.0)

    # min 값이 max 값보다 커지는 경우 (매우 작은 contour 등에서 발생 가능성) 처리
    if y_min_norm >= y_max_norm or x_min_norm >= x_max_norm:
        # 이 경우, 면적이 0인 박스로 처리하거나, 입력 rect가 매우 작았음을 의미
print(f"Warning: Degenerate bbox calculated in rect_to_minmax: y_min={y_min_norm}, y_max={y_max_norm}, x_min={x_min_norm}, x_max={x_max_norm}. Returning zero bbox.")
        return [0.0, 0.0, 0.0, 0.0]


    bbox_normalized = [y_min_norm, x_min_norm, y_max_norm, x_max_norm]
    return bbox_normalized

# 함수 호출 시에는 item['image'].shape 을 전달합니다.
pred_bbox = rect_to_minmax(rect, item['image'].shape)
pred_bbox
```

```python
def get_iou(boxA, boxB):
    y_min = max(boxA[0], boxB[0])
    x_min= max(boxA[1], boxB[1])
    y_max = min(boxA[2], boxB[2])
    x_max = min(boxA[3], boxB[3])
    
    interArea = max(0, x_max - x_min) * max(0, y_max - y_min)
    boxAArea = (boxA[2] - boxA[0]) * (boxA[3] - boxA[1])
    boxBArea = (boxB[2] - boxB[0]) * (boxB[3] - boxB[1])
    iou = interArea / float(boxAArea + boxBArea - interArea)

    return iou
```
